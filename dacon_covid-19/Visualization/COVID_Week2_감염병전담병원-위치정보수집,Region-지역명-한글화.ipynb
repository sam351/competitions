```python
# 지역별 감염병 전담 병원 수
str_list = str_raw.split()
r_dict = {}

for idx, item in enumerate(str_list):
    if item in '0123456789':
        r_dict[ str_list[idx-1] ] = item
print(r_dict)
```

```python
# 구글 api로 병원 주소 및 위경도 생성
location = hospitals[0]
URL = f'https://maps.googleapis.com/maps/api/geocode/json?address={location}&key={api_key}&language=ko'
response = requests.get(URL)
data = response.json()
data

lat = data['results'][0]['geometry']['location']['lat']
lng = data['results'][0]['geometry']['location']['lng']
addr = data['results'][0]['formatted_address']

print(location, lat, lng, addr)
print()
pprint.pprint(data)
```

```python
province_tag_list = []
for key in r_dict.keys():
    if key == '합계':
        continue
    province_tag_list.extend( [key] * eval(r_dict[key]) )
    pass
print(len(province_tag_list))
print(province_tag_list)
```

```python
# 구글 api로 병원별 주소 및 위경도 수집
lat_list = []
lng_list = []
addr_list = []
for idx, tmp_location in enumerate(hospitals):
    tmp_location = province_tag_list[idx] + ' ' + tmp_location
    
    URL = f'https://maps.googleapis.com/maps/api/geocode/json?address={tmp_location}&key={api_key}&language=ko'
    response = requests.get(URL)
    data = response.json()
    
    if data['status'] == 'ZERO_RESULTS':
        print(f'>>> Error : No result for {tmp_location} (index : {idx})')
        continue
    
    tmp_lat = data['results'][0]['geometry']['location']['lat']
    tmp_lng = data['results'][0]['geometry']['location']['lng']
    tmp_addr = data['results'][0]['formatted_address']
    
    lat_list.append(tmp_lat)
    lng_list.append(tmp_lng)
    addr_list.append(tmp_addr)
    pass

print('>>> Finished\n')
print(hospitals[-1], lat_list[-1], lng_list[-1], addr_list[-1])
```

```python
# 결측 데이터 추가
tmp_location = '충청남도 연기군 남면 나성리 566'
URL = f'https://maps.googleapis.com/maps/api/geocode/json?address={tmp_location}&key={api_key}&language=ko'
response = requests.get(URL)
data = response.json()

tmp_lat = data['results'][0]['geometry']['location']['lat']
tmp_lng = data['results'][0]['geometry']['location']['lng']
tmp_addr = data['results'][0]['formatted_address']

lat_list.insert(29, tmp_lat)
lng_list.insert(29, tmp_lng)
addr_list.insert(29, tmp_addr)
```

```python
# 병원별 주소 및 위경도 데이터프레임 생성 및 저장
hospital_df = pd.DataFrame({'name':hospitals, 'lat':lat_list, 'lng':lng_list, 'address':addr_list})
hospital_df.to_csv('감염병전담병원_위치정보.csv', encoding='cp949')

hospital_df.loc[25:30]
```

## Region.csv의 지역명 한글화

```python
import pandas as pd

import requests
from kaggle_secrets import UserSecretsClient
api_key = UserSecretsClient().get_secret('google_api_key')
```

```python
# 지역정보 데이터 로드
region_df = pd.read_csv('../input/coronavirusdataset/Region.csv')
region_df['addr'] = region_df.apply(lambda row : row['province'] + ' ' + row['city'] if row['province'] != row['city'] else row['province'], axis=1)
region_df
```

```python
# 구글 api로 한글 지역명 및 위경도 수집
lat_list = []
lng_list = []
addr_list = []
for idx, tmp_location in enumerate(region_df.addr):
    URL = f'https://maps.googleapis.com/maps/api/geocode/json?address={tmp_location}&key={api_key}&language=ko'
    response = requests.get(URL)
    data = response.json()
    
    if data['status'] == 'ZERO_RESULTS':
        print(f'>>> Error : No result for {tmp_location} (index : {idx})')
        continue
    
    tmp_lat = data['results'][0]['geometry']['location']['lat']
    tmp_lng = data['results'][0]['geometry']['location']['lng']
    tmp_addr = data['results'][0]['formatted_address']
    
    lat_list.append(tmp_lat)
    lng_list.append(tmp_lng)
    addr_list.append(tmp_addr)
    pass

print('>>> Finished\n')

# 마지막 데이터 확인
print(tmp_location, lat_list[-1], lng_list[-1], addr_list[-1])
```

```python
# 데이터프레임에 수집된 데이터 추가
region_df['google_addr_kor'] = addr_list
region_df['google_lat'] = lat_list
region_df['google_lng'] = lng_list

region_df
```

```python
# 이상치 확인
display(region_df[ (region_df.google_addr_kor.str[:4]!='대한민국') | (region_df.google_addr_kor.str.len()>15) ])

# 이상치 수정(세종시, 창원시, 한국)
region_df.loc[81, 'google_addr_kor'] = '대한민국 세종특별자치시'
region_df.loc[236, 'google_addr_kor'] = '대한민국 경상남도 창원시'
region_df.loc[243, 'google_addr_kor'] = '대한민국'

# 이상치 없어진 것 확인
display(region_df[ (region_df.google_addr_kor.str[:4]!='대한민국') | (region_df.google_addr_kor.str.len()>15) ])
```

```python
# 시/도 한글명, 시/군/구 한글명 분리 저장
region_df['google_addr_kor'] = region_df['google_addr_kor'].str.replace('대한민국 ', '')
region_df['province_kor'] = region_df['google_addr_kor'].map(lambda x : x.split()[0])
region_df['city_kor'] = region_df['google_addr_kor'].map(lambda x : x.split()[0] if len(x.split())==1 else x.split()[1])

# 불필요 컬럼 제거 후 저장 및 확인
region_df = region_df.drop(columns=['addr', 'google_addr_kor'])
region_df.to_csv('Region_translated.csv', index=False)
region_df
```
